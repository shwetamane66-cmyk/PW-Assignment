**Q2.	How does broadcasting work in NumPy?**

**Broadcasting** is a feature in NumPy that allows arrays of different shapes to perform arithmetic operations without explicitly copying data. NumPy automatically expands the smaller array to match the shape of the larger array when possible.

**Rules of Broadcasting**
- If the arrays have the same shape, operations are performed element-wise.
- If the shapes are different, NumPy compares the dimensions from right to left.
- Two dimensions are compatible if:
  - They are equal, or
  - One of them is 1.
- If the dimensions are not compatible, NumPy raises a ValueError.

```python
# Example
import numpy as np

# Create a 2D array
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

# Create a 1D array
b = np.array([10, 20, 30])

# Broadcasting addition
result = arr + b

print("Original Array:")
print(arr)

print("\nArray to Broadcast:")
print(b)

print("\nResult:")
print(result)
```

**Summary**

- Broadcasting allows NumPy to perform operations on arrays of different shapes efficiently without creating unnecessary copies of data, making computations faster and more memory-efficient.

**Q4.	Explain the use of the groupby() method in Pandas.**

The **groupby()** method in Pandas is used to group data based on one or more columns and perform operations such as sum, mean, count, minimum, maximum, and other aggregate functions on each group.

**Uses of groupby() :**
- Groups data by one or more columns.
- Performs aggregate functions like sum(), mean(), count(), min(), and max().
- Helps summarize and analyze large datasets.
- Commonly used in data analysis and reporting.

```python
# Example
import pandas as pd

# Create a DataFrame
data = {
    'Department': ['HR', 'IT', 'HR', 'IT', 'Sales'],
    'Salary': [30000, 50000, 35000, 55000, 40000]
}

df = pd.DataFrame(data)

# Group by Department and calculate the average salary
result = df.groupby('Department')['Salary'].mean()

print(result)
```

**Explanation**
- df.groupby('Department') groups the rows by the Department column.
- ['Salary'] selects the Salary column.
- .mean() calculates the average salary for each department.

**Q8.	What does the term "vectorized operation" mean in NumPy?**

A **vectorized operation** in NumPy is the process of performing an operation on all elements of an array at once without using explicit loops. Vectorized operations are faster, more efficient, and require less code than traditional Python loops.

**Advantages of Vectorized Operations**
- Faster execution than using for loops.
- Simpler and cleaner code.
- Memory efficient.
- Performs element-wise operations on entire arrays.

```python
# Example
import numpy as np

# Create a NumPy array
arr = np.array([1, 2, 3, 4, 5])

# Vectorized operation
result = arr * 2

print("Original Array:", arr)
print("After Multiplication:", result)
```

**Explanation**
- arr * 2 multiplies every element in the array by 2 in a single operation.
- No for loop is required.
- NumPy performs the computation internally, making it much faster than iterating through each element manually.

**Reasons Why NumPy Slicing is Faster**
- Creates a view instead of a copy (in most cases), reducing memory usage.
- Stores data in contiguous memory, enabling faster access.
- Uses optimized C implementations for array operations.
- Performs operations more efficiently than Python lists.

```python
import numpy as np

# Create a NumPy array
arr = np.array([10, 20, 30, 40, 50])

# Slice the array
print(arr[1:4])
```

**Explanation**

```python
import pandas as pd

# Create a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Marks': [85, 70, 95, 60, 88]
}

df = pd.DataFrame(data)

# Set the threshold
threshold = 80

# Filter rows where Marks are greater than the threshold
filtered_df = df[df['Marks'] > threshold]

# Display the filtered DataFrame
print(filtered_df)
```

**Q10.	Create a histogram using Seaborn to visualize a distribution.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = [12, 15, 18, 20, 22, 22, 25, 27, 30, 30, 32, 35, 40]

# Create a histogram
sns.histplot(data, bins=6, kde=True)

# Add title and labels
plt.title("Histogram of Sample Data")
plt.xlabel("Values")
plt.ylabel("Frequency")

# Display the plot
plt.show()
```

**Q11.	Perform matrix multiplication using NumPy.**

```python
import numpy as np

# Create two matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Perform matrix multiplication
result = np.matmul(A, B)

# Display the matrices and result
print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nMatrix Multiplication (A × B):")
print(result)
```

**Q13. Create a 3D scatter plot using Plotly**

```python
import plotly.express as px
import pandas as pd

# Create a sample DataFrame
data = {
    'X': [1, 2, 3, 4, 5],
    'Y': [5, 4, 3, 2, 1],
    'Z': [10, 15, 12, 20, 25]
}

df = pd.DataFrame(data)

# Create a 3D scatter plot
fig = px.scatter_3d(
    df,
    x='X',
    y='Y',
    z='Z',
    title='3D Scatter Plot using Plotly'
)

# Display the plot
fig.show()
```
