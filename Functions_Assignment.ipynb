**Q1. What is the difference between a function and a method in Python?**

Answer :     

A **function** is a block of reusable code that performs a specific task and is defined independently of any class or object. It can be called directly by using its name whereas
A **method** is a function that is defined inside a class and is associated with an object of that class. Methods are called using an object or class and can access or modify the object's data.**bold text**








**Example**

```python
# Function
def greet(name):
    return "Hello, " + name

print(greet("Alice"))

# Method
class Person:
    def greet(self):
        return "Hello, Alice"

p = Person()
print(p.greet())
```

**Conclusion** : A function is an independent block of code, whereas a method is a function that belongs to a class and is used to perform operations on objects of that class.

**Q5.  What are iterators in Python and how do they differ from iterables?**

Answer:

In Python, iterables and iterators are used to access elements of a collection one at a time.

**Iterable:** An iterable is an object that contains a collection of elements and can be traversed one by one. Examples include lists, tuples, strings, dictionaries, and sets.

**Iterator:** An iterator is an object that keeps track of the current position during iteration and provides the next element when requested. An iterator is created from an iterable using the iter() function.

```python
# EXample of Iterable
my_list = [1, 2, 3]

# Example of Iterator
my_list = [1, 2, 3]
it = iter(my_list)

print(next(it))  # 1
print(next(it))  # 2
print(next(it))  # 3
```

Difference Between Iterable and Iterator:

Iterable
*  A collection of elements that can be iterated over.
*  Uses the __iter__() method.
*  Cannot directly provide the next element.
*  Examples: List, Tuple, String, Set, Dictionary.

Iterator
*  An object that produces elements one at a time.
*  Uses both __iter__() and __next__() methods.
*  Can provide the next element using next().
*  Object returned by iter().

**Q6.  Explain the concept of generators in Python and how they are defined.**

Answer:

* A generator in Python is a special type of function that produces a sequence of values one at a time instead of returning all values at once. Generators are memory-efficient because they generate values only when needed.
* Generators use the yield keyword instead of the return statement. When a generator function is called, it returns a generator object without executing the function immediately.

**Defining a Generator:** A generator is defined like a normal function, but it contains one or more yield statements

```python
# Example
def count_numbers():
    yield 1
    yield 2
    yield 3

gen = count_numbers()

print(next(gen))
print(next(gen))
print(next(gen))
```

How Generators Work

* The yield statement pauses the function and saves its current state.
* When next() is called again, execution resumes from where it stopped.
* This continues until all values have been generated.
Generator Expression

**Generator Expression :** Generators can also be created using generator expressions, which are similar to list comprehensions.

```python
squares = (x * x for x in range(5))

for num in squares:
    print(num)
```

**Advantages of Generators**


*   Memory Efficient – Values are generated on demand.
*   Faster Execution – Useful for processing large datasets.
*   Simpler Code – Makes iterator creation easier.

**Conclusion:** Generators are special functions that use the yield keyword to generate values one at a time. They are useful for handling large amounts of data efficiently and provide a convenient way to create iterators in Python.

**Q7.What are the advantages of using generators over regular functions?**

Answer:

**Advantages of Using Generators Over Regular Functions**

Generators offer several advantages compared to regular functions, especially when working with large datasets or sequences.

1. Memory Efficiency : Generators produce values one at a time instead of storing all values in memory. This makes them suitable for handling large amounts of data.

2. Faster Execution : Since values are generated only when needed, generators can start producing output immediately without waiting for the entire sequence to be created.

3. Improved Performance : Generators reduce memory usage and processing overhead, leading to better performance in many applications.

4. Easy Implementation of Iterators : Generators automatically implement the iterator protocol, making it easier to create iterators without defining __iter__() and __next__() methods.

5. Ability to Handle Infinite Sequences : Generators can generate an unlimited sequence of values without consuming excessive memory.

**Q8. What is a lambda function in Python and when is it typically used?**

- A lambda function is an anonymous, single-expression function defined using the lambda keyword. It is commonly used for short, simple operations, especially when passing functions as arguments to other functions.

Lambda functions are typically used when:

*  A simple function is needed for a short period of time.
*  A function is required as an argument to another function.

**Q9. . Explain the purpose and usage of the `map()` function in Python.**

Answer:

- The map() function in Python is used to apply a specified function to each item of an iterable (such as a list, tuple, or set) and returns a map object containing the results.
- It helps perform operations on all elements of a collection without using explicit loops.

Syntax

```python
map(function, iterable)
```

- function: The function to be applied to each element.
- iterable: The collection of elements to process.

```python
# Example
numbers = [1, 2, 3, 4, 5]

squares = map(lambda x: x * x, numbers)

print(list(squares))
```

```python
# Using a Regular Function with map()
def square(x):
    return x * x

numbers = [1, 2, 3, 4]

result = map(square, numbers)

print(list(result))
```

**Advantages of map()**

1.Reduces the need for explicit loops.

2.Makes code shorter and more readable.

3.Efficient for applying the same operation to multiple elements.
Works well with lambda functions.

**Q.10  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

Answer:

Difference Between map(), reduce(), and filter() Functions in Python

map(), filter(), and reduce() are built-in functions in Python used for processing iterables such as lists and tuples.


1. map() Function

The map() function applies a specified function to every element in an iterable.

```python
numbers = [1, 2, 3, 4]

result = map(lambda x: x * 2, numbers)

print(list(result))
```

2. filter() Function

The filter() function selects only those elements that meet a specified condition.

```python
numbers = [1, 2, 3, 4, 5, 6]

result = filter(lambda x: x % 2 == 0, numbers)

print(list(result))
```

3. reduce() Function

The reduce() function, available in the functools module, applies a function cumulatively to the elements of an iterable and returns a single value.

```python
from functools import reduce

numbers = [1, 2, 3, 4]

result = reduce(lambda x, y: x + y, numbers)

print(result)
```

**Conclusion :**

- map() transforms every element in an iterable.

- filter() selects elements based on a condition.

- reduce() combines all elements into a single result.

These functions are widely used in Python for concise and efficient data processing.

**Q1. . Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.**

```python
def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

# Example
numbers = [10, 15, 20, 25, 30]
print("Sum of even numbers:", sum_even_numbers(numbers))
```

**Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.**

```python
def is_prime(num):
    if num <= 1:
        return False

    for i in range(2, num):
        if num % i == 0:
            return False

    return True

# Example
number = 97

if is_prime(number):
    print(number, "is a Prime Number")
else:
    print(number, "is not a Prime Number")
```

**Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.**

```python
class Fibonacci:
    def __init__(self, n):
        self.n = n          # Number of terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration

        fib_num = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1

        return fib_num

# Example
fib = Fibonacci(10)

for num in fib:
    print(num, end=" ")
```

**Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.**

```python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example
for value in powers_of_two(5):
    print(value)
```

**Q8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.**

**Using a Lambda Function to Sort a List of Tuples by the Second Element**

```python
# List of tuples
students = [("John", 85), ("Alice", 92), ("Bob", 78), ("David", 88)]

# Sort by the second element (marks)
sorted_students = sorted(students, key=lambda x: x[1])

print(sorted_students)
```

```python
sorted_students = sorted(students, key=lambda x: x[1], reverse=True)
print(sorted_students)
```

**Q9.  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.**

```python
# List of temperatures in Celsius
celsius = [0, 10, 20, 30, 40]

# Convert Celsius to Fahrenheit using map()
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("Celsius:", celsius)
print("Fahrenheit:", fahrenheit)
```

**Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string.**

```python
# Function to remove vowels
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    result = "".join(filter(lambda ch: ch not in vowels, text))
    return result

# Example
string = "Hello World"
print("Original String:", string)
print("After Removing Vowels:", remove_vowels(string))
```
