Q6. Why are lists mutable, but tuples are immutable?

Answer :

- Lists are mutable because you can change their contents after making them you can add, remove, or replace items whereas tuples are immutable because once you make them, you can’t change anything inside — no adding, removing, or modifying.

See below example :



```python
# List - can change
fruits = ["apple", "banana"]
fruits[0] = "mango"
print(fruits)
```

```python
# Tuple - can't change
colors = ("red", "blue")
colors[0] = "green"
print(colors) #If you print this code, it will give an error.
```

Q14. Describe a scenario where a while loop is more suitable than a for loop?
Answer :
- A while loop is more suitable when you don’t know in advance how many times you’ll need to loop — you just need to keep running until a certain condition becomes true.

Example scenario:
A program keeps asking a user for their password until they enter the correct one.

```python
correct_password = "python123"
user_input = ""

while user_input != correct_password:
    user_input = input("Enter password: ")

print("Access granted!")
```

Q6. Write a function to demonstrate the behavior of mutable and immutable arguments.

```python
""" Answer """
def modify_values(num, my_list):
    # Trying to change the integer
    num = num + 10

    # Trying to change the list
    my_list.append(100)

    print("Inside function - num:", num)
    print("Inside function - my_list:", my_list)


# Immutable argument (integer)
x = 5

# Mutable argument (list)
y = [1, 2, 3]

print("Before function call - x:", x)
print("Before function call - y:", y)

modify_values(x, y)

print("\nAfter function call - x:", x)
print("After function call - y:", y)
```

Q9. Write a Python program to convert user input from string to integer, float, and boolean types.

```python
""" Answer """
user_input = input("Enter a value: ")

print("\nOriginal Input (string):", user_input)

# Convert to integer
try:
    int_value = int(user_input)
    print("As Integer:", int_value)
except ValueError:
    print("Cannot convert to Integer")

# Convert to float
try:
    float_value = float(user_input)
    print("As Float:", float_value)
except ValueError:
    print("Cannot convert to Float")

# Convert to boolean
# In Python, bool("anything") → True, except empty string → False
bool_value = bool(user_input)
print("As Boolean:", bool_value)
```

Q10. Write code to demonstrate type casting with list elements.

```python
""" Answer """

str_list = [1, 2, 3, 4]

print("Original List:", str_list)

# Convert each element to integer
int_list = [int(x) for x in str_list]
print("As Integers:", int_list)

# Convert each element to float
float_list = [float(x) for x in str_list]
print("As Floats:", float_list)

# Convert each element to boolean
bool_list = [bool(x) for x in str_list]
print("As Booleans:", bool_list)
```

Q13. Write a Python program to find the sum of all even numbers between 1 and 50.

Q14.  Write a program to reverse a string using a while loop.

```python
""" Answer """
text = input("Enter a string: ")

reversed_text = ""
i = len(text) - 1   # start from the last index

while i >= 0:
    reversed_text += text[i]
    i -= 1

print("Reversed string:", reversed_text)
```

Q15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
""" Answer """
num = int(input("Enter a number: "))

factorial = 1
i = 1

while i <= num:
    factorial *= i
    i += 1

print("Factorial of", num, "is:", factorial)
```
