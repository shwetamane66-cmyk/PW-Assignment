**Q8.	What is a hash table, and how does it relate to dictionaries in Python?**

A **hash table** is a data structure that stores data in key-value pairs and allows very fast insertion, deletion, and lookup of values.

It works by using a hash function to convert a key into a unique index (location) where the value is stored.

How a Hash Table Works :
1. A key is provided.
2. A hash function generates a hash value (index).
3. The value is stored at that index.
4. When the key is used again, the hash function finds the value quickly.


```python
# Example of hash table
student = {
    "name": "Shweta",
    "age": 27
}

print(student["name"])

# In this example, Python uses a hash table internally to quickly find the value associated with the key "name".
```

Relationship Between Hash Tables and Dictionaries
- A dictionary in Python is implemented using a hash table.
- Keys are hashed to determine where values are stored in memory.
- This allows dictionaries to access data very efficiently.

Key Features of Python Dictionaries
- Store data as key-value pairs.
- Keys must be unique and hashable (e.g., strings, numbers, tuples containing immutable elements).
- Provide fast access, insertion, and deletion of data.

```python
# Example of Adding and Accessing Data
employee = {
    "id": 101,
    "name": "Alice"
}

employee["department"] = "HR"

print(employee["department"])
```

A **hash table** is a data structure that stores data using keys and values for fast retrieval. Python dictionaries are built on hash tables, which is why dictionary operations such as searching, adding, and updating data are very efficient.

**Q17.	Describe the time complexity of accessing elements in a dictionary.**

In Python, accessing an element in a dictionary using its key has an average time complexity of O(1) (constant time).

Why Is It O(1)?

Python dictionaries are implemented using hash tables. When you provide a key:

1. Python computes the key's hash value.
2. The hash value determines where the data is stored.
3. Python directly retrieves the value from that location.

This allows very fast access regardless of the dictionary's size.

```python
# Example
student = {
    "name": "Shweta",
    "age": 27,
    "course": "Python"
}

print(student["name"])
```

Accessing "name" takes approximately the same amount of time whether the dictionary contains 3 items or 3 million items.

Common Dictionary Operations
1. Access a value (dict[key]) - O(1) average
2. Insert a key-value pair - O(1) average
3. Update a value - O(1) average
4. Delete a key-value pair - O(1) average
5. Search for a key (key in dict) - O(1) average


Worst-Case Scenario

In rare cases, multiple keys may produce the same hash value (hash collision). When many collisions occur, access time can degrade to O(n).

Summary
- Dictionary element access is O(1) on average.
- This efficiency comes from the use of hash tables.
- In rare collision-heavy situations, the worst-case complexity can be O(n).

In simple terms:
Python dictionaries can usually find a value instantly using its key, which makes them much faster than searching through a list.

**Q19.	Why are dictionaries considered unordered, and how does that affect data retrieval?**

Dictionaries are considered unordered (historically) because they store data using a hash table, where elements are placed based on a hash value of the key, not by a fixed.

**Why They Are Considered Unordered?**
- Each key in a dictionary is processed through a hash function.
- The hash determines where the key-value pair is stored in memory.
- This storage is based on hash values, not insertion position.
- So, there is no guaranteed “first” or “last” element in terms of logical ordering (especially in older Python versions).

**Important Note (Modern Python)**

From Python 3.7 onward, dictionaries preserve insertion order, but they are still conceptually based on hashing, not indexed ordering like lists.

**How This Affects Data Retrieval**

1. Retrieval is Based on Keys, Not Position

   You do not access dictionary elements by index.

```python
student = {
    "name": "John",
    "age": 20,
    "course": "Python"
}

print(student["age"])
```

2. Fast Access (O(1))
  
   Because of hashing:
- Python directly jumps to the memory location.
- No need to search sequentially like in lists.

3. No Reliance on Position

   Even if insertion order is preserved in modern Python, you should not depend on it for logic like:
- “first item means most important”
- “second item means next step”

  Instead, always use keys for retrieval.

```python
# Example Showing No Index Access
student = {"name": "John", "age": 20}

# student[0]  ❌ Not allowed
# This will raise an error because dictionaries are not accessed by index.
```

Summary:
- Dictionaries are based on hash tables, not sequential order.
- They are considered unordered in theory because storage depends on hashing, not position.
- Data is retrieved using keys, not indexes.
- Retrieval is very fast (O(1)) and independent of order.
