**Q5. What are dunder methods in Python?**

Dunder methods (short for Double UNDERscore methods) are special methods in Python that begin and end with double underscores (__). They are also called magic methods or special methods.

These methods allow you to define how objects of a class behave with built-in operations such as printing, addition, comparison, and object creation.

Common Dunder Methods and It's purpose:

1. __init__()- Constructor; initializes an object
2. __str__()- Returns a string representation of an object
3. __add__()- Defines behavior for the + operator
4. __len__()- Defines behavior for len()
5. __eq__()- Defines behavior for the == operator
6. __del__()- Destructor; called when an object is deleted

````python
# Example: __init__() and __str__()
class Person:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Person: {self.name}"

p = Person("Shweta")
print(p)
````

````python
# Example 2: __add__() (Operator Overloading)
class Number:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return Number(self.value + other.value)

    def __str__(self):
        return str(self.value)

n1 = Number(10)
n2 = Number(20)

print(n1 + n2)
````

**Why Are Dunder Methods Useful?**
- Customize the behavior of objects.
- Support operator overloading (+, -, *, etc.).
- Make objects work with built-in functions like print(), len(), and comparisons.
- Improve code readability and flexibility.

**Q7.	What is polymorphism in OOP?**

Polymorphism is an OOP concept that allows the same method or interface to behave differently for different objects.

The word polymorphism means "many forms". It enables one method name to perform different actions depending on the object that calls it.

````python
# Example: Method Overriding
class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def sound(self):
        print("Bark!")

class Cat(Animal):
    def sound(self):
        print("Meow!")

dog = Dog()
cat = Cat()

dog.sound()
cat.sound()
````

````python
# Runtime Polymorphism
class Bird:
    def fly(self):
        print("Bird flies")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies high")

class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly")

birds = [Sparrow(), Penguin()]

for bird in birds:
    bird.fly()
````

Benefits of Polymorphism
- Improves code flexibility.
- Makes programs easier to extend.
- Reduces code duplication.
- Allows a common interface for different types of objects.

Summary

Polymorphism is the ability of different objects to respond to the same method call in their own way, allowing one interface to represent multiple behaviors.

**Q8.	How is encapsulation achieved in Python?**

Encapsulation is the process of bundling data (attributes) and methods (functions) together in a class and restricting direct access to some of the object's data.

In Python, encapsulation is achieved using access modifiers:

- Public members: Accessible from anywhere.
- Protected members (_variable): Intended for internal use within the class and its subclasses.
- Private members (__variable): Cannot be accessed directly from outside the class.

````python
# Example of Encapsulation
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private attribute

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient balance")

    def get_balance(self):
        return self.__balance

# Create object
account = BankAccount(1000)

account.deposit(500)
account.withdraw(200)

print("Balance:", account.get_balance())
````

Benefits of Encapsulation
- Protects data from unauthorized access.
- Prevents accidental modification of data.
- Improves code security and maintainability.
- Provides controlled access through methods.

Encapsulation in Python is achieved by using classes and private attributes (__variable) along with public methods to control access to data, ensuring data security and integrity.

**Q10.	What are class and static methods in Python?**

Class Methods and Static Methods in Python

Both class methods and static methods belong to a class rather than a specific object, but they differ in how they access class data.

| Feature                   | Class Method               | Static Method            |
| ------------------------- | -------------------------- | ------------------------ |
| Decorator                 | `@classmethod`             | `@staticmethod`          |
| First Parameter           | `cls` (class reference)    | No special parameter     |
| Access Class Variables    | Yes                        | No                       |
| Access Instance Variables | No (unless passed)         | No                       |
| Purpose                   | Work with class-level data | Utility/helper functions |


1. Class Method

- A class method receives the class itself (cls) as its first argument and can access or modify class variables.

````python
class Person:
    count = 0

    def __init__(self):
        Person.count += 1

    @classmethod
    def get_count(cls):
        return cls.count

p1 = Person()
p2 = Person()

print(Person.get_count())
````

2. Static Method

- A static method does not receive self or cls. It behaves like a regular function placed inside a class.

````python
class MathOperations:

    @staticmethod
    def add(a, b):
        return a + b

print(MathOperations.add(10, 5))
````

````python
# Example
class Employee:
    company = "ABC Ltd"

    @classmethod
    def company_name(cls):
        return cls.company

    @staticmethod
    def calculate_tax(salary):
        return salary * 0.1

print(Employee.company_name())
print(Employee.calculate_tax(50000))
````

Summary
- Class Method (@classmethod): Works with class-level data and receives cls as the first parameter.
- Static Method (@staticmethod): Does not access class or instance data and is used for utility functions.
- Use class methods when you need to work with class variables.
- Use static methods when the method is related to the class but does not need access to class or object data.

**Q11.	What is method overloading in Python?**

What is Method Overloading in Python?

Method overloading is the ability to define multiple methods with the same name but different parameters in a class.

Unlike languages such as Java or C++, Python does not support traditional method overloading. If you define multiple methods with the same name, the last definition overrides the previous ones.

````python
# Example
class Demo:
    def show(self):
        print("No arguments")

    def show(self, name):
        print("Hello", name)

obj = Demo()
obj.show("Shweta")
````

How Python Achieves Similar Behavior

Python uses:

- Default arguments
- Variable-length arguments (*args, **kwargs)

````python
class Demo:
    def show(self, name=None):
        if name is None:
            print("No arguments")
        else:
            print("Hello", name)

obj = Demo()
obj.show()
obj.show("Shweta")
````

````python
# Using *args
class Calculator:
    def add(self, *args):
        return sum(args)

calc = Calculator()

print(calc.add(10, 20))
print(calc.add(10, 20, 30))
print(calc.add(10, 20, 30, 40))
````

Method overloading means having multiple methods with the same name but different parameter lists. Python does not support it directly, but similar functionality can be achieved using default parameters and variable-length arguments (*args, **kwargs).

**Q12.	What is method overriding in OOP?**

Method overriding occurs when a child class provides its own implementation of a method that is already defined in its parent class.

The overridden method in the child class has the same name and parameters as the method in the parent class.

````python
# Example
# Parent class
class Animal:
    def speak(self):
        print("Animal makes a sound")

# Child class
class Dog(Animal):
    def speak(self):
        print("Bark!")

# Create objects
animal = Animal()
dog = Dog()

animal.speak()
dog.speak()
````

How It Works
- Animal defines the method speak().
- Dog inherits from Animal and overrides speak().
- When dog.speak() is called, Python executes the Dog version of the method instead of the Animal version.

````python
# Using super()
# You can call the parent class method from the child class using super():
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        super().speak()
        print("Bark!")

dog = Dog()
dog.speak()
````

Benefits of Method Overriding
- Provides specialized behavior in child classes.
- Supports runtime polymorphism.
- Increases code flexibility and reusability.
- Allows customization of inherited methods.

Summary- Method overriding is the process of redefining a parent class method in a child class with the same name and parameters, allowing the child class to provide its own specific behavior.

**Q13.	What is a property decorator in Python?**

What is a Property Decorator in Python?

The @property decorator allows you to define a method that can be accessed like an attribute. It is commonly used to implement encapsulation by controlling access to private variables.

Instead of calling a method with parentheses, you can access it like a normal attribute.

````python
# Example
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

# Create object
p = Person("Shweta")

# Access like an attribute
print(p.name)
````

````python
# Property with Getter and Setter
class Person:
    def __init__(self, age):
        self._age = age

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, value):
        if value >= 0:
            self._age = value
        else:
            print("Age cannot be negative")

p = Person(25)

print(p.age)

p.age = 30
print(p.age)
````

````python
# Property with Deleter
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

    @name.deleter
    def name(self):
        del self._name

p = Person("Alice")
del p.name
````

Benefits of @property
- Provides controlled access to attributes.
- Supports encapsulation.
- Allows validation before setting values.
- Makes code cleaner and more readable.
- Lets you change internal implementation without changing how the attribute is accessed.

Summary :

The @property decorator converts a method into a read-only attribute and is often used with getter, setter, and deleter methods to provide controlled access to class data while maintaining encapsulation.

**Q15.	What is an abstract class in Python?**

An abstract class is a class that cannot be instantiated (cannot create objects directly) and is used as a blueprint for other classes.

It can contain:
- Abstract methods (methods without implementation).
- Regular methods (methods with implementation).

In Python, abstract classes are created using the ABC (Abstract Base Class) module from the abc package.

````python
from abc import ABC, abstractmethod

# Abstract class
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

# Derived class
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

# Create object
circle = Circle(5)
print("Area:", circle.area())
````

Why Can't We Create an Object of an Abstract Class?

````python
shape = Shape()

# This will raise an error because Shape contains an abstract method that has not been implemented.
````

````python
# Example with Multiple Child Classes
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        print("Moo")

class Sheep(Animal):
    def sound(self):
        print("Baa")

cow = Cow()
sheep = Sheep()

cow.sound()
sheep.sound()
````

Benefits of Abstract Classes
- Enforces a common interface for all child classes.
- Promotes code consistency.
- Supports abstraction by hiding implementation details.
- Improves code maintainability and design.

Key Points
- Use ABC to create an abstract class.
- Use @abstractmethod to define abstract methods.
- Abstract classes cannot be instantiated.
- Child classes must implement all abstract methods.

**Q18.	What is multiple inheritance in Python?**

Multiple inheritance - It is a feature in Python where a child class inherits from more than one parent class. This allows the child class to access the attributes and methods of all its parent classes.

````python
class Parent1:
    pass

class Parent2:
    pass

class Child(Parent1, Parent2):
    pass
````

````python
# Parent class 1
class Father:
    def skill1(self):
        print("Driving")

# Parent class 2
class Mother:
    def skill2(self):
        print("Cooking")

# Child class
class Child(Father, Mother):
    pass

# Create object
c = Child()

c.skill1()
c.skill2()
````

How It Works :
- Child inherits from both Father and Mother.
- Therefore, it can access methods from both parent classes.
- This promotes code reuse and combines functionality from multiple classes.

Method Resolution Order (MRO)

If multiple parent classes have methods with the same name, Python follows the Method Resolution Order (MRO) to determine which method to call.

````python
class A:
    def show(self):
        print("Class A")

class B:
    def show(self):
        print("Class B")

class C(A, B):
    pass

obj = C()
obj.show()
````

- Combines functionalities from different sources.
- Reduces code duplication.
- Increases flexibility in program design.

Disadvantages
- Can make code more complex.
- May lead to ambiguity when parent classes have methods with the same name.
- Requires understanding of MRO to avoid confusion.

Summary- Multiple inheritance is a mechanism in Python where a class inherits from two or more parent classes, allowing it to access and combine their attributes and methods. Python uses Method Resolution Order (MRO) to resolve conflicts when multiple parent classes contain methods with the same name.

**Q20.	What is the significance of the 'super()' function in Python?**

The super() function is used to call methods and constructors of a parent class from a child class.

It is mainly used in inheritance to avoid rewriting code and to access the functionality of the parent class.

Why Use super()?
- Reuse code from the parent class.
- Initialize parent class attributes.
- Call overridden parent methods.
- Support multiple inheritance through Method Resolution Order (MRO).

````python
# Example 1: Calling the Parent Class Constructor
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, roll_no):
        super().__init__(name)  # Call parent constructor
        self.roll_no = roll_no

s = Student("Shweta", 101)

print(s.name)
print(s.roll_no)
````

````python
# Example 2: Calling a Parent Class Method
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        super().speak()
        print("Dog barks")

dog = Dog()
dog.speak()
````

Benefits of super()

**Q22.	What is the difference between @staticmethod and @classmethod in Python?**

Both @staticmethod and @classmethod are methods that belong to a class rather than an instance, but they differ in how they interact with the class.

| Feature                   | `@staticmethod`        | `@classmethod`                          |
| ------------------------- | ---------------------- | --------------------------------------- |
| Decorator                 | `@staticmethod`        | `@classmethod`                          |
| First Parameter           | No special parameter   | `cls` (class reference)                 |
| Access Class Variables    | No                     | Yes                                     |
| Access Instance Variables | No                     | No (unless an object is passed)         |
| Purpose                   | Utility/helper methods | Methods that work with class-level data |


@staticmethod

A static method does not receive self or cls. It behaves like a normal function placed inside a class.

````python
class MathOperations:

    @staticmethod
    def add(a, b):
        return a + b

print(MathOperations.add(10, 5))
````

@classmethod

A class method receives the class itself (cls) as its first parameter and can access or modify class variables.

````python
class Student:
    school = "ABC School"

    @classmethod
    def get_school(cls):
        return cls.school

print(Student.get_school())
````

````python
# Example Showing the Difference
class Employee:
    company = "Tata Motors"

    @classmethod
    def company_name(cls):
        return cls.company

    @staticmethod
    def calculate_bonus(salary):
        return salary * 0.10

print(Employee.company_name())
print(Employee.calculate_bonus(50000))
````

When to Use Each?

Use @staticmethod when: The method does not need access to class or instance data. It performs a utility task related to the class.

Use @classmethod when: The method needs to access or modify class variables. You want to create alternative constructors.

Summary

- @staticmethod: Does not receive self or cls; used for utility functions related to the class.
- @classmethod: Receives cls as the first parameter and can access or modify class-level data.
- Use static methods for independent helper functions and class methods when working with the class itself.

**Q23.	How does polymorphism work in Python with inheritance?**

**Polymorphism** in Python with inheritance allows a child class to provide its own implementation of a method that is defined in the parent class. When the method is called, Python executes the version that belongs to the actual object, not the reference type.

This is achieved through method overriding.

````python
# Example
# Parent class
class Animal:
    def sound(self):
        print("Animal makes a sound")

# Child class
class Dog(Animal):
    def sound(self):
        print("Bark!")

# Child class
class Cat(Animal):
    def sound(self):
        print("Meow!")

# Create objects
animals = [Dog(), Cat()]

for animal in animals:
    animal.sound()
````

How It Works :
- Animal defines the method sound().
- Dog and Cat inherit from Animal.
- Both child classes override the sound() method.
- When animal.sound() is called, Python determines the object's actual type at runtime and executes the appropriate method.

This behavior is called runtime polymorphism.

````python
# Another Example Using a Function
class Shape:
    def draw(self):
        print("Drawing a shape")

class Circle(Shape):
    def draw(self):
        print("Drawing a circle")

class Rectangle(Shape):
    def draw(self):
        print("Drawing a rectangle")

def display(shape):
    shape.draw()

display(Circle())
display(Rectangle())
````

Benefits of Polymorphism with Inheritance
- Flexibility: One interface can handle different object types.
- Code Reusability: Common functionality is inherited from the parent class.
- Extensibility: New child classes can be added without changing existing code.
- Maintainability: Reduces code duplication and simplifies updates.

**Q24.	What is method chaining in Python OOP?**

Method chaining is a technique in which multiple methods are called on the same object in a single statement.

It is achieved by having each method return the object itself (self), allowing the next method to be called immediately.

````python
# Example
class Calculator:
    def __init__(self, value=0):
        self.value = value

    def add(self, num):
        self.value += num
        return self

    def subtract(self, num):
        self.value -= num
        return self

    def multiply(self, num):
        self.value *= num
        return self

    def display(self):
        print("Result:", self.value)
        return self

# Method chaining
calc = Calculator()
calc.add(10).subtract(2).multiply(5).display()
````

How It Works
- add(10) updates the value and returns self.
- Since self is returned, subtract(2) can be called immediately.
- The process continues with multiply(5) and display().

Equivalent code without method chaining:

````python
calc = Calculator()

calc.add(10)
calc.subtract(2)
calc.multiply(5)
calc.display()
````

Advantages of Method Chaining
- Makes code shorter and cleaner.
- Improves readability.
- Reduces the need for temporary variables.
- Provides a fluent interface for object manipulation.

Disadvantage
- Long chains can become difficult to debug if an error occurs in one of the methods.

Summary : Method chaining is a technique where methods return self, allowing multiple method calls to be linked together in a single statement. It helps create cleaner, more readable, and fluent code.

**Q2.	Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.**

````python
from abc import ABC, abstractmethod
import math

# Abstract class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Circle class
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

# Rectangle class
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Create objects
circle = Circle(5)
rectangle = Rectangle(4, 6)

# Display areas
print("Area of Circle:", circle.area())
print("Area of Rectangle:", rectangle.area())
````

**Q5.	Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.**

````python
class BankAccount:
    def __init__(self, balance=0):
        self.__balance = balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: ₹{amount}")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: ₹{amount}")
        else:
            print("Insufficient balance.")

    def check_balance(self):
        print(f"Current Balance: ₹{self.__balance}")

# Create an account
account = BankAccount(1000)

# Perform operations
account.check_balance()
account.deposit(500)
account.withdraw(300)
account.check_balance()
````
