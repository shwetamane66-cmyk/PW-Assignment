**Q1. What is the difference between interpreted and compiled languages?**

**Difference Between Interpreted and Compiled Languages**

---



Interpreted Language:
- The code is executed line by line by an interpreter.
- No separate executable file is created.
- Execution is generally slower.
- Errors are reported one at a time during execution.
- Easier to test and debug.
- Example: Python, JavaScript, Ruby

Compiled Language
- The entire source code is translated into machine code before execution.
- A separate executable file is created after compilation.
- Execution is generally faster.
- Most errors are detected during compilation before the program runs.
- Debugging may require recompiling the program after changes.
- Example: C, C++, Go


**Key Difference**
- Interpreted languages execute code line by line, making development easier but execution slower.
- Compiled languages convert the entire program into machine code before execution, making them generally faster.

**Q9.	What is the difference between multithreading and multiprocessing?**

Difference Between Multithreading and Multiprocessing

---

**Multithreading**
- Uses multiple threads within a single process.
- Threads share the same memory.
- Best suited for I/O-bound tasks (e.g., file handling, network requests).
- Has lower memory usage and less overhead.
- If one thread crashes, it may affect the entire process.
- Uses the threading module.

**Multiprocessing**
- Uses multiple processes, each with its own memory space.
- Each process has separate memory.
- Best suited for CPU-bound tasks (e.g., mathematical calculations, image processing).
- Uses more memory and has higher overhead.
- One process crashing usually does not affect other processes.
- Uses the multiprocessing module.

```python
# Example: Multithreading
import threading

def display():
    print("Thread is running")

t = threading.Thread(target=display)
t.start()
t.join()
```

```python
# Example: Multiprocessing
from multiprocessing import Process

def display():
    print("Process is running")

p = Process(target=display)
p.start()
p.join()
```

**Key Difference :**
- Multithreading runs multiple threads within the same process and is ideal for I/O-bound tasks.
- Multiprocessing runs multiple processes independently and is ideal for CPU-bound tasks because it can take advantage of multiple CPU cores.

**Q12.	What are the basic steps involved in exception handling in Python?**

**Exception handling** is used to detect and handle runtime errors so that the program continues to execute without crashing.

**Basic Steps :**
1. Write the risky code inside the try block
   - Place the code that may cause an exception in the try block.
2. Handle the exception using the except block
   - If an error occurs, the corresponding except block is executed.
3. Use the else block (optional)
   - The else block executes only if no exception occurs in the try block.
4. Use the finally block (optional)
   - The finally block always executes, whether an exception occurs or not. It is commonly used for cleanup tasks such as closing files.

```python
# Syntax
try:
    # Code that may raise an exception

except ExceptionType:
    # Handle the exception

else:
    # Executes if no exception occurs

finally:
    # Always executes
```

**Explanation**

- The except block catches and handles the exception.

**Q15.	How does Python's garbage collection system work?**

Python's garbage collection (GC) system automatically frees memory by removing objects that are no longer in use. This helps prevent memory leaks and ensures efficient memory management.

**How It Works**
1. **Reference Counting :**
- Every object in Python has a reference count.
- When a new reference to an object is created, the reference count increases.
- When a reference is removed, the reference count decreases.
If the reference count becomes 0, the object's memory is automatically released.
2. **Garbage Collector :**
- Python also has a garbage collector that detects and removes cyclic references (objects that reference each other but are no longer accessible).
- This ensures that memory used by such objects is also reclaimed.

```python
# Example
# Create a list object
numbers = [10, 20, 30]

# Another reference to the same object
x = numbers

# Remove one reference
del numbers

print(x)

# Remove the last reference
del x

print("Memory can now be reclaimed.")
```

**Explanation**
- numbers and x both refer to the same list.
- del numbers removes one reference, but the object still exists because x refers to it.
- del x removes the last reference.
- Since the reference count becomes 0, Python can automatically free the memory.

**Q18.	What is the difference between os.fork() and multiprocessing in Python?**

Both os.fork() and the multiprocessing module are used to create
new processes, but they differ in portability, ease of use, and features.



---



**os.fork()**
- Creates a child process by duplicating the parent process.
- Available only on Unix/Linux systems (not supported on Windows).
- Low-level system call.
- Parent and child continue executing from the same point after the fork
- Requires manual process management.

**multiprocessing**
- Creates and manages separate processes using a high-level API.
- Works on Windows, Linux, and macOS.
- High-level Python module.
- Processes run a specified target function.
- Provides easy process management, communication, and synchronization.

```python
# Example using os.fork()
import os

pid = os.fork()

if pid == 0:
    print("This is the child process.")
else:
    print("This is the parent process.")
```

```python
# Example using multiprocessing
from multiprocessing import Process

def display():
    print("This is the child process.")

p = Process(target=display)
p.start()
p.join()

print("Parent process finished.")
```

**Summary**
- os.fork() is a low-level system call that creates a copy of the current process and is available only on Unix/Linux.
- multiprocessing is a high-level, cross-platform module that makes creating and managing processes simpler and is generally the recommended approach in Python.

**Q18.	How would you implement a basic logging setup that logs to a file with rotation after lMB?**

```python
import logging
from logging.handlers import RotatingFileHandler

# Create a rotating file handler
handler = RotatingFileHandler(
    "app.log",          # Log file name
    maxBytes=1024 * 1024,  # 1 MB
    backupCount=3          # Keep 3 backup files
)

# Set log format
formatter = logging.Formatter(
    "%(asctime)s - %(levelname)s - %(message)s"
)
handler.setFormatter(formatter)

# Create logger
logger = logging.getLogger()
logger.setLevel(logging.INFO)
logger.addHandler(handler)

# Log some messages
logger.info("Application started.")
logger.warning("This is a warning message.")
logger.error("This is an error message.")

print("Logs have been written to app.log")
```

**Q21.	Write a Python program that reads a file and prints the number of occurrences of a specific word.**

```python
# Read a file and count occurrences of a specific word

word = input("Enter the word to search: ")

with open("sample.txt", "r") as file:
    content = file.read().lower()      # Convert to lowercase
    count = content.count(word.lower())

print(f"The word '{word}' occurs {count} time(s).")
```

**Q22.	How can you check if a file is empty before attempting to read its contents?**

```python
import os

filename = "sample.txt"

# Check if the file is empty
if os.path.getsize(filename) == 0:
    print("The file is empty.")
else:
    with open(filename, "r") as file:
        content = file.read()
        print("File Contents:")
        print(content)
```

**Q23.	Write a Python program that writes to a log file when an error occurs during file handling.**

```python
import logging

# Configure logging
logging.basicConfig(
    filename="error.log",
    level=logging.ERROR,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

try:
    # Try to open a file that may not exist
    with open("sample.txt", "r") as file:
        content = file.read()
        print(content)

except Exception as e:
    # Log the error to the log file
    logging.error(f"An error occurred: {e}")
    print("An error occurred. Check 'error.log' for details.")
```
